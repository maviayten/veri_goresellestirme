```python
import pandas as pd
import numpy as np

np.random.seed(0)

data = {
    "Yaş": np.random.randint(20, 65, 200),
    "Gelir": np.random.randint(3000, 20000, 200),
    "Kategori": np.random.choice(["A", "B", "C", "D", "E"], 200),
    "Zaman": pd.date_range(start="2023-01-01", periods=200, freq='D'),
    "Değer": np.random.randint(1, 100, 200).cumsum(),
    "Puan": np.random.randn(200) * 1000 + 5000
}

df = pd.DataFrame(data)

df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df['Yaş_Aralığı'] = pd.cut(df['Yaş'], bins=[20, 30, 40, 50, 60, 70], labels=['20-30', '30-40', '40-50', '50-60', '60+'])

plt.figure(figsize=(6, 4))

sns.countplot(x='Kategori',
              data=df,
              hue='Yaş_Aralığı',
              palette='coolwarm',
              order=df['Kategori'].value_counts().index,
              dodge=True,
              linewidth=2,
              edgecolor='black',
              saturation=0.8,
              width=0.7)

plt.title('Kategori ve Yaş Aralıklarına Göre Dağılım (Seaborn)', fontsize=10)
plt.xlabel('Kategori', fontsize=10)
plt.ylabel('Frekans', fontsize=10)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Yaş Aralığı', loc='upper right', fontsize=7, title_fontsize=7)
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.7)

plt.show()
```
